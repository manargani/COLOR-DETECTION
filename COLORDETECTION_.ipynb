```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/81b54a1d6b4fdefb8d868ccf9c6bcf32--php (1).jpg')
```

```python
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
imageRGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,8))
```

```python
plt.imshow(imageRGB)
```

```python
imageHSV=cv2.cvtColor(imageRGB,cv2.COLOR_RGB2HSV)
```

```python
lower = np.array([25,150,50])
upper = np.array([30,255,255])
```

```python
mask = cv2.inRange(imageHSV , lower , upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res = cv2.bitwise_and(imageRGB, imageRGB ,mask=mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res)
```

```python
lower = np.array([130, 0, 0])
upper = np.array([200, 255, 255])
```

```python
maskred=cv2.inRange(imageHSV,lower,upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(maskred)
```

```python
res1 = cv2.bitwise_and(imageRGB, imageRGB ,mask=maskred)
```

```python
plt.figure(figsize=(20,8))
```

```python
plt.imshow(res1)
```
